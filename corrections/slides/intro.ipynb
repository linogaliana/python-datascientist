## Objectifs du cours

-   Pouvoir utiliser `Python` pour l’<span class="orange">**ensemble du processus de valorisation**</span> des données :
    -   Récupérer et structurer des données ;
    -   Manipuler ;
    -   Visualiser ;
    -   Modéliser.

. . .

-   Comprendre comment `Python` sert dans une <span class="orange">**démarche de recherche ou de production**</span> ;

. . .

-   S’initier à la question de la <span class="orange">**reproductibilité** </span> (*open-source*, *open-data*…) ;

. . .

-   Découvrir la <span class="orange">**pratique moderne**</span> de `Python` dans le monde de la *data-science* ;

. . .

-   Découvrir qu’on peut <span class="orange">**faire des choses sympa**</span> ✨ avec `Python` .

    -   Le projet doit être **documenté**

. . .

    -   Discutez avec votre chargé de TD pour avoir un avis ;

. . .

>
> -   Exigence de <span class="orange">**reproductibilité**</span>
>     -   Projet utilisant des **sources *open-data*** ou scrappant des sites publics ;
>     -   Code doit pouvoir être **répliqué** par chargé TD ;
>     -   <span class="orange">**Pas un projet Kaggle !**</span>
